```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
import cv2
image = cv2.imread('/car_image.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
if image is None:
    print("Error: Image not found!")
    exit()
```

```python
#1.Image negative
image_negative = 255 - image
```

```python
## 2. Contrast Stretching
r_min = np.min(image)
r_max = np.max(image)

contrast_stretched = ((image - r_min) / (r_max - r_min)) * 255
contrast_stretched = contrast_stretched.astype(np.uint8)
```

```python
# 3. Bit Plane Slicing
bit_planes = []
for i in range(8):
    bit_plane = (image >> i) & 1
    bit_planes.append(bit_plane * 255)
```

```python
# 4. Gray Level Slicing
lower = 100
upper = 150

gray_slicing = np.zeros_like(image)
gray_slicing[(image >= lower) & (image <= upper)] = 255
```

```python
plt.figure(figsize=(15, 8))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(image, cmap='rainbow')
plt.title('Rainbow Image')
plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(12, 10))

plt.subplot(3, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 2)
plt.title("Image Negative")
plt.imshow(image_negative, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 3)
plt.title("Contrast Stretching")
plt.imshow(contrast_stretched, cmap='gray')
plt.axis('off')

for i in range(6):
    plt.subplot(3, 3, i + 4)
    plt.title(f"Bit Plane {i}")
    plt.imshow(bit_planes[i], cmap='gray')
    plt.axis('off')

plt.subplot(3, 3, 9)
plt.title("Gray Level Slicing")
plt.imshow(gray_slicing, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))

plt.subplot(3, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='rainbow')
plt.axis('off')

plt.subplot(3, 3, 2)
plt.title("Image Negative")
plt.imshow(image_negative, cmap='rainbow')
plt.axis('off')

plt.subplot(3, 3, 3)
plt.title("Contrast Stretching")
plt.imshow(contrast_stretched, cmap='rainbow')
plt.axis('off')

for i in range(6):
    plt.subplot(3, 3, i + 4)
    plt.title(f"Bit Plane {i}")
    plt.imshow(bit_planes[i], cmap='rainbow')
    plt.axis('off')

plt.subplot(3, 3, 9)
plt.title("Gray Level Slicing")
plt.imshow(gray_slicing, cmap='rainbow')
plt.axis('off')

plt.tight_layout()
plt.show()
```
